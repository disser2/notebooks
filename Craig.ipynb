# Basic Imports

```python
# Standard Libs
import numpy as np
import pandas as pd

# Machine Learning Libs
from sklearn import linear_model, svm, tree, cluster, manifold, datasets

# Plots
import pylab as pl
%matplotlib inline
```

# Beispiel: Clustering

## Rohdaten

```python
X, y = datasets.make_blobs(n_samples=1000, centers=4, random_state=10)
pl.scatter(X[:, 0], X[:, 1])
```

## Wir berechnen automatisch Cluster von Punkten

```python
db = cluster.DBSCAN(eps=2, min_samples=5).fit(X)
colors = [int(i % 23) for i in db.labels_]
pl.scatter(X[:, 0], X[:, 1], c=colors)
```
